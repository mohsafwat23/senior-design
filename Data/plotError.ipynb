```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import os
import glob


dir = str(os.path.abspath(os.curdir))

files = glob.glob(dir + "/*.csv")
print(files)

for file in files:
    with open(file, 'r') as csvfile:
        plots = csv.reader(csvfile, delimiter=',')
        x = []
        y = []
        count = 0
        for row in plots:

            t_point = float(row[0])
            y_point = float(row[1])
            if y_point < 2.4239:
                if count == 0:
                    t_init = float(row[0])
                count+=1
                x.append(t_point-t_init)
                y.append(y_point)
    legend_name = file.replace(dir + "/", "")
    legend_name = legend_name.replace(".csv", "")

    plt.plot(x, y, label=legend_name)

plt.xlabel('Time [s]')
plt.ylabel('Error [m]')
plt.title('Tuning PID')
plt.legend()
plt.show()
```
